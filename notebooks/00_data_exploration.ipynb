## Load Sample Data

For now, we'll create sample nutrition documents for testing.

```python
# Sample nutrition knowledge base
nutrition_documents = [
    "Protein is essential for muscle growth and repair. The recommended daily intake is 0.8g per kilogram of body weight for sedentary adults, but athletes may need 1.2-2.0g/kg.",
    "Carbohydrates are the body's primary energy source. Complex carbs like whole grains provide sustained energy, while simple carbs provide quick energy but can cause blood sugar spikes.",
    "Healthy fats from sources like avocados, nuts, and olive oil are important for hormone production and nutrient absorption. Aim for 20-35% of daily calories from fat.",
    "Vitamins and minerals are essential micronutrients. Vitamin D supports bone health, B vitamins support energy metabolism, and calcium is crucial for bone strength.",
    "Water intake recommendations vary, but the general guideline is 8 glasses per day. However, needs depend on activity level, climate, and individual factors. Drink enough to keep urine pale yellow.",
    "Fiber is important for digestive health. Aim for 25-30g daily from sources like whole grains, legumes, vegetables, and fruits. Fiber helps with satiety and blood sugar control.",
    "Weight loss requires a calorie deficit. A deficit of 500 calories per day typically results in 0.5kg (1 lb) weight loss per week. However, metabolism varies by individual.",
    "Balanced meals should include proteins, carbohydrates, and healthy fats. This combination helps maintain stable blood sugar and keeps you feeling full longer.",
    "Sodium intake should be limited to less than 2,300mg per day. Excess sodium can increase blood pressure. Choose fresh foods over processed ones when possible.",
    "Sugar consumption should be minimized. Added sugars provide empty calories and can lead to weight gain and increased disease risk. Limit to less than 25g daily for women and 36g for men.",
]

# Create a DataFrame
df = pd.DataFrame({
    'doc_id': range(len(nutrition_documents)),
    'content': nutrition_documents,
    'word_count': [len(doc.split()) for doc in nutrition_documents],
    'char_count': [len(doc) for doc in nutrition_documents]
})

print(f"Total documents: {len(df)}")
print(f"\nDataFrame:")
print(df.head())
```

## Data Statistics


```python
print(f"Average words per document: {df['word_count'].mean():.1f}")
print(f"Average characters per document: {df['char_count'].mean():.1f}")
print(f"\nWord count statistics:")
print(df['word_count'].describe())
print(f"\nDocument topics:")
keywords = ['protein', 'carbohydrate', 'fat', 'vitamin', 'water', 'fiber', 'weight', 'meal', 'sodium', 'sugar']
for keyword in keywords:
    count = sum(1 for doc in nutrition_documents if keyword.lower() in doc.lower())
    print(f"  {keyword}: {count} documents")
```

## Save Sample Data


```python
# Save to CSV for later use
os.makedirs('../data/processed', exist_ok=True)
df.to_csv('../data/processed/nutrition_docs.csv', index=False)
print("Sample data saved to data/processed/nutrition_docs.csv")
```
