## Build RAG Prompt Example


```python
context = """[Source 1] Protein is essential for muscle growth and repair. The recommended daily intake is 0.8g per kilogram of body weight for sedentary adults, but athletes may need 1.2-2.0g/kg.

[Source 2] Balanced meals should include proteins, carbohydrates, and healthy fats. This combination helps maintain stable blood sugar and keeps you feeling full longer."""

query = "How much protein do I need daily?"

rag_prompt = PromptManager.build_rag_prompt(query, context, prompt_type='standard', few_shot=True)

print("Complete RAG Prompt:")
print("="*60)
print(rag_prompt)
```
